```python
#3.

The ArithmeticError class is a base class for exceptions that occur during arithmetic operations. 
Here are two examples of exceptions derived from ArithmeticError, along with explanations and examples:

1. ZeroDivisionError:
import logging

logging.basicConfig (filename = 'error.log', level = logging.ERROR)

try:
    result = 10 / 0  
except ZeroDivisionError as e:
    logging.error("Error:", e)
    
2. OverflowError:

import sys

import logging

logging.basicConfig (filename = 'error2.log', level = logging.ERROR)

try:
    large_number = sys.maxsize + 1 
except OverflowError as e:
   logging.error ("Error:", e)
```

```python
#4.

The LookupError class is a base class for exceptions related to dictionary or sequence lookups that fail. 
It's used to handle situations where an attempt to access a value or element using a key or index is unsuccessful. 

Example:
    
1. KeyError:
This exception is raised when a dictionary is accessed with a key that doesn't exist in the dictionary.

my_dict = {'a': 1, 'b': 2, 'c': 3}

import logging

logging.basicConfig (filename = 'error.log', level = logging.ERROR)

try:
    value = my_dict['d']  
except KeyError as e:
    logging.error("Error:", e)

2. IndexError:
This exception is raised when attempting to access an index that is out of range in a sequence (like a list or a tuple).

my_list = [10, 20, 30]

import logging

logging.basicConfig (filename = 'error.log', level = logging.ERROR)

try:
    value = my_list[3]  
except IndexError as e:
    logging.error("Error:", e)
```

```python
#5.

ImportError is a base class for all exceptions related to importing modules. 
It's raised when there is a problem with the import statement, such as when the specified module cannot be found, or there's an issue with the module's content.

logging.basicConfig (filename = 'error.log', level = logging.ERROR)
try:
    import non_existent_module  
except ImportError as e:
    logging.error("Import Error:", e)
    
ModuleNotFoundError is a specific subclass of ImportError. 
It's introduced in Python 3.6 to provide a more precise error message when a module cannot be found during an import operation. 
```
